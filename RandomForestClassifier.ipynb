```
# importing necessary python libraries
import numpy as np
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

# setting up train and test files
train = pd.read_excel('TrainingSet.xlsx')
test = pd.read_excel('TestSet1.xlsx')

# setting up train and test data sets
X_train = train.iloc[:, :4]
y_train = train.plant
X_test = test.iloc[:, :4]
y_test = test.plant
```

```
# Random Forest Classifier with 100 estimators i.e 100 Trees and random state = 123
classifier = RandomForestClassifier(n_estimators=100, random_state = 123)

# Fitting the model
classifier.fit(X_train, y_train)
# predicting the model
y_test = classifier.predict(X_test)

# setting up the plant column in the test data set
test['plant'] = y_test

# saving the contents of the data set into an excel file and setting the index to False
# so that the indexes of data sets may not interfere with the excel file when written
saver = test.to_excel('Test100.xlsx', index = False)

# showing the results of the predictions
reader = pd.read_excel('Test100.xlsx')
reader
```
